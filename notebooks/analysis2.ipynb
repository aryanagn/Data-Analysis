```python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("../data/raw/ks-projects-201801.csv")

#make a few graphs, this one is not a good example yet.
sb.countplot(y = 'state', data=df)
plt.title("State of Projects")
```

This is a pretty basic plot to see the general counts of failed versus succesful products, generally more failed than where succseful.

```python

df['launched']= pd.to_datetime(df['launched']).dt.date
df['deadline']= pd.to_datetime(df['deadline']).dt.date
df['project_time'] =  (df['deadline'] -df['launched']).astype('timedelta64[D]').astype(int)
j= df.project_time
sb.countplot(x = j,  hue = df.state, dodge = False)
plt.title("Project running time importance")
```

The days are hard to see, so will be smarter to do per week on the x axis, but generally thus plot gives a rough look into the most common times for projects and when they are most succesful. It looks to be around the 21-30 day mark where there is a big spike, this is the most common project time, however more fail than succeed. Generally less than 3 weeks sees more success than failures. Will be easier to track once 'state' is cleaned up to just success or pass (task 2)

```python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("../data/raw/ks-projects-201801.csv")
dk = df.copy()
dk['name_length'] = dk['name'].str.len()
dk['pledged'] = dk['pledged']
vis = sb.scatterplot(data =dk, y = 'name_length', x= 'pledged', hue = 'state')
plt.title("Name length importance")
```

Eexcluding outliers that wil be removed in the date cleaning process, we can see generally that projects
with a name of more than 60 characters fail to recieve any funding. The sweet spot for character length of name is between 20-60.

```python
df['launched']= pd.to_datetime(df['launched']).dt.date
df['launched_month'] = df.launched.apply(lambda x: x.month)
sb.countplot(x= df['launched_month'], hue = df.state)
plt.title("Launch month vs success rate")
```

Based on these trends, it appears theat July is the most popular month to launch a project, however it is not the most succesful while also having the most fails. The most sucesful month is March. There is a much lower project count in December, and generally most are unsuccseful.

# Task 3 Method Chaining
Method chaning and move to python file

```python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd


def load_and_process(path_csv):
    
    df1 = (
         pd.read_csv( path_csv)
        .drop(['category','currency','country', 'pledged', 'goal', 'usd pledged'], axis=1)
        .rename(columns={'main_category': 'category', 'usd_pledged_real': 'pledged', 'usd_goal_real': 'goal'})
    )
    df2 = (
        df1.loc[df1['state'].isin(
            ['successful', 'failed'])]
    )
  
    return df2

load_and_process("../data/raw/ks-projects-201801.csv").head()

```
